```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')


plt.figure(figsize=(8,5))
plt.title('Gas Prices Over Time (in USD)', 
          fontdict={'fontweight':'bold','fontsize':22})


plt.plot(gas.Year , gas.USA , 'b.-' , label ="USA" )
plt.plot(gas.Year , gas.Canada , 'r.-',label ="Canada")
plt.plot(gas.Year , gas['South Korea'] ,'g.-', label ="South Korea")
plt.plot(gas.Year , gas.Australia ,'y.-', label ="Australia")

#ways to plot many values at once
# countries_to_look_at =['Australia' ,"USA","Canada","South Korea"]
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.' ,label = country)



plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_prices_graph.png', dpi=300)

plt.show()
```

### FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

#### Histogram

```python
bins =[40,50,60,70,80,90,100]
plt.hist(fifa.Overall , bins = bins , color = "#84f542")  


plt.ylabel('Numbers of Players')
plt.xlabel('Skill level')

plt.title('Distribution of Player skills in FIFA 2018')

plt.savefig('Player Skill Distribution FIFA 2018')
plt.show()
```

### Pie Charts

```python
left  = (fifa['Preferred Foot'] == 'Left').sum()
right = (fifa['Preferred Foot'] == 'Right').sum()

color =['#38f2d0','#760ff5']
plt.pie([left,right], labels =['Left','Right'] , colors= color, autopct='%.2f %%')

plt.title('Foot Preference Of FIFA Players')

plt.savefig("Preferred Foot of FIFA Players")
plt.show()
```

```python
fifa.Weight =[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight[0]
```

```python
plt.style.use('ggplot')

light = (fifa.Weight < 125).sum()
light_medium = ((fifa.Weight >= 125) & (fifa.Weight < 150)).sum()
medium = ((fifa.Weight >= 150) & (fifa.Weight < 175)).sum()
medium_heavy = ((fifa.Weight >= 175) & (fifa.Weight < 200)).sum()
heavy = (fifa.Weight >= 200).sum()

weights = [light, light_medium, medium, medium_heavy, heavy] 
label =['under 125', '125-150','150-175','175-200' , 'Over 200']
explode  =(.5,0.2,0,0,.5)

plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.pie(weights, labels = label, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona  = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs =fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels =['FC Barcelona','Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid,revs], labels = labels, patch_artist = True, medianprops={'linewidth':2})
colors = [ ]
for box in boxes['boxes']:
    #set edge color 
    box.set(color = '#4286f4' , linewidth =2)
    
    #change fill color
    box.set(facecolor = '#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA overall Rating')

plt.show()
```
